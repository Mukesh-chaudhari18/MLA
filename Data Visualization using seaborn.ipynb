````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Load sample dataset
tips = sns.load_dataset('tips')
tips.head()
````

## 1️⃣ Scatter Plot
**Purpose:** Used to visualize relationships between two numerical variables.

````python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
````

### 🔹 Customization Options
- `hue`: Color points based on a category
- `size`: Change point sizes based on a variable
- `style`: Different marker styles for categories

````python
sns.scatterplot(x='total_bill', y='tip', hue='sex', size='size',markers="Sex", data=tips)
plt.title('Customized Scatter Plot')
plt.show()
````

## 2️⃣ Line Plot
**Purpose:** Used to show trends over time.

````python
sns.lineplot(x='size', y='tip', data=tips)
plt.title('Line Plot: Tip by Party Size')
plt.show()
````

````python
sns.lineplot(x='month',y='passengers',data=flight)
plt.show()
````

### 🔹 Customization Options
- `hue`: Different colors for categories
- `markers`: Markers at data points
- `dashes`: Different line styles

````python
sns.lineplot(x='size', y='tip', hue='sex', style='sex', markers=True, dashes=True, data=tips)
plt.title('Customized Line Plot')
plt.show()
````

## 3️⃣ Bar Plot
**Purpose:** Used to compare categorical data.

````python
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Bar Plot: Total Bill by Day')
plt.show()
````

````python
flight.info()
````

### 🔹 Customization Options
- `hue`: Different colors for categories
- `palette`: Custom color palettes
- `ci`: Confidence interval (set to None to remove)

````python
sns.barplot(x='day', y='total_bill', hue='day', palette='coolwarm', ci=None, data=tips)
plt.title('Customized Bar Plot')
plt.show()
````

## 4️⃣ Box Plot
**Purpose:** Used to show the distribution and detect outliers.

````python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot: Total Bill by Day')
plt.show()
````

### 🔹 Customization Options
- `hue`: Color boxes by category
- `palette`: Change color scheme
- `showfliers`: Hide outliers

````python
sns.boxplot(x='day', y='total_bill', hue='sex', palette='pastel', showfliers=False, data=tips)
plt.title('Customized Box Plot')
plt.show()
````

## 5️⃣ Heatmap
**Purpose:** Used to visualize correlations and matrix data.

````python
corr = tips.corr(numeric_only=True)
````

````python
corr
````

````python

sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
````

## 6️⃣ Histogram
**Purpose:** Shows the distribution of numerical data.

````python
sns.histplot(tips['total_bill'], bins=20, kde = True)
plt.title('Histogram: Total Bill Distribution')
plt.show()
````

### 🔹 Customization Options
- `bins`: Number of bins
- `kde`: Whether to show Kernel Density Estimate
- `color`: Change bar colors

````python
sns.histplot(tips['total_bill'], bins=15, kde=True, color='green')
plt.title('Customized Histogram')
plt.show()
````

## 7️⃣ Violin Plot
**Purpose:** Combines box plot and KDE plot to show distribution.

````python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot: Total Bill by Day')
plt.show()
````

````python
sns.violinplot(x='smoker', y='total_bill', data=tips)
plt.title('Violin Plot: Total Bill by Day')
plt.show()
````

### 🔹 Customization Options
- `hue`: Color by category
- `split`: Split the violin into multiple categories
- `palette`: Change color scheme

````python
sns.violinplot(x='day', y='total_bill', hue='sex', palette='muted', data=tips)
plt.title('Customized Violin Plot')
plt.show()
````

## 8️⃣ Count Plot
**Purpose:** Shows the count of observations in categorical data.

````python
sns.countplot(x='day', data=tips,fill=True)
plt.title('Count Plot: Number of Orders by Day')
plt.show()
````

### 🔹 Customization Options
- `hue`: Group counts by another category
- `palette`: Change color scheme

````python
sns.countplot(x='day', hue='smoker', palette='coolwarm', data=tips)
plt.title('Customized Count Plot')
plt.show()
````

## 9️⃣ Pair Plot
**Purpose:** Shows pairwise relationships between numerical variables.

````python
sns.pairplot(tips)
plt.title('Pair Plot of Dataset')
plt.show()
````

### 🔹 Customization Options
- `hue`: Color by category
- `kind`: Use different plots like 'reg' for regression
- `diag_kind`: Change diagonal plots (e.g., 'hist', 'kde')

````python
sns.pairplot(tips, diag_kind='kde', palette='coolwarm')
plt.title('Customized Pair Plot')
plt.show()
````
